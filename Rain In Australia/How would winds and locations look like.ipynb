```python
preprocessor = Preprocessor('median', 'mode', 
                            roll_cols=standard_roll_cols, 
                            roll_strategies=standard_roll_strategies,
                            roll_period=7)
data = preprocessor.preprocess(train, visualize=True)
```

```python
location = 'Sydney'
locations = preprocessor.locations.copy(deep=True)
selected_columns = [column for column in data.columns if 'RainToday_' in column]
total_rains = data.groupby('Location')['RainTomorrow'].sum()
total_rains.index.name = 'index'
locations = locations.merge(total_rains.rename('Total_rains').reset_index(), left_on='Location_reduced', right_on='index').drop('index', axis=1)
relative_raintoday = data.loc[(data.RainTomorrow == 1) & (data.Location == location), selected_columns].sum(axis=0)
relative_raintoday.index = relative_raintoday.index.to_series().apply(lambda txt: txt[10:])
locations = locations.merge(relative_raintoday.rename('relative_RainToday').reset_index(), left_on='Location_reduced', right_on='index').drop('index', axis=1)
```
